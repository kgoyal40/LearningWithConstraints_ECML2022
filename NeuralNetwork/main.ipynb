```python
class NN:
    def __init__(self):
        self.n_hidden_layers = 1
        self.models = {}
    
    def learn_linear_model(self, X, y):
        pass
    
    def learn_layer_model(self, X, y, layer_id=0):
        X_output = pd.DataFrame()
        for j in range(len(self.models[layer_id])):
            self.models[layer_id][j].learn_without_sade(X, y, M=[0], learner='fumalik')
            X_output[j] = [v if v > 0 else 0 for v in self.models[layer_id][j].predict(X, y)[3]]
        return X_output
        
    def learn(self, X, y, n_hidden_layers=[[2]]):
        if len(n_hidden_layers) > 1:
            raise VaueError('invalid input: n_hidden_layers')
        
        self.models[0] = [LinearClassificationLearnerLoan()]*n_hidden_layers[0][0]
        self.models[1] = [LinearClassificationLearnerLoan()]
        
        X_new = self.learn_layer_model(X, y, layer_id=0)
        self.learn_layer_model(X_new, y, layer_id=1)
        
        
```
