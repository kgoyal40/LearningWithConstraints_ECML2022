```python
import pandas as pd
```

```python
full_data = pd.read_csv('Family Income and Expenditure.csv', sep=',')
```

```python
full_data.head()
```

```python
partial_data = pd.read_csv('family_income.csv', sep=',')
```

```python
partial_data
```

```python
expenses_types = [c for c in full_data.columns if 'Expenditure' in c]
expenses_types
```

```python
All_Food_Expenses = [sum([full_data[e][i] for e in expenses_types[:7]])/10000 for i in range(full_data.shape[0])]
Going_Out_Expenses = [sum([full_data[e][i] for e in expenses_types[7:9]])/10000 for i in range(full_data.shape[0])]
Shopping_Expenses = [sum([full_data[e][i] for e in expenses_types[9:11]])/10000 for i in range(full_data.shape[0])]
Necessary_Expenses = [sum([full_data[e][i] for e in expenses_types[11:16]])/10000 for i in range(full_data.shape[0])]
Miscellaneous_Expenses = [sum([full_data[e][i] for e in expenses_types[16:]])/10000 for i in range(full_data.shape[0])]

print(expenses_types[:7], expenses_types[7:9], expenses_types[9:11], expenses_types[11:16], expenses_types[16:])
```

```python
partial_data.columns[:-5] + ['All_Food_Expenses', 'Going_Out_Expenses', 'Shopping_Expenses', 'Necessary_Expenses', 'Miscellaneous_Expenses']
```

```python
cols = ['Number of Television', 'Number of Car, Jeep, Van',
       'Number of Stove with Oven/Gas Range', 'House Floor Area', 'House Age',
       'Number of bedrooms', 'Electricity', 'Total Number of Family members',
       'Total number of family members employed', 'Household Head Sex',
       'Household Head Age', 'Agricultural Household indicator', 'Total Household Income']
data = full_data[cols]
data['Total Household Income'] = [d/10000 for d in data['Total Household Income']]
```

```python
data['All_Food_Expenses'] = All_Food_Expenses
data['Going_Out_Expenses'] = Going_Out_Expenses
data['Shopping_Expenses'] = Shopping_Expenses
data['Necessary_Expenses'] = Necessary_Expenses
data['Miscellaneous_Expenses'] = Miscellaneous_Expenses
```
