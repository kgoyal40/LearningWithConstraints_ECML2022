```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python

delta = 0.1
x = np.arange(-2.0, 2.0, delta)
y = np.arange(-2, 2, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(0.3*(-X**2 - 3*Y**2))
# Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
# Z = (Z1 - Z2) * 2

fig, ax = plt.subplots()
plt.rcParams["figure.figsize"] = (10, 10)
# plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams["axes.titleweight"] = "bold"
matplotlib.rcParams['font.size'] = 35
plt.rcParams.update({'figure.autolayout': True})

CS = ax.contour(X, Y, Z1, linewidths=2)
# ax.set_title('SaDe: Satisfiability Descent')

pac = mpatches.Wedge([-1.75, 0.75], 0.5, theta1=270, theta2=360, facecolor='xkcd:grey', edgecolor='grey', ls = '--', alpha=0.3)
plt.gca().add_patch(pac)

plt.annotate(s='', xy=(-1.75, 0.8), xytext=(-1.75 + 0.5, 0.80), arrowprops=dict(arrowstyle='<->, head_width=0.1'))
plt.text(-1.75 + 0.2, 0.85, r'$\alpha$', fontsize=40)

plt.annotate(s='', xy=(-1.8, 0.78), xytext=(-1.8, 0.8 - 0.55), arrowprops=dict(arrowstyle='<->, head_width=0.1'))
plt.text(-1.8 - 0.2, 0.5, r'$\alpha$', fontsize=40)

plt.plot(-1.75, 0.75, 'k', markersize=10, marker = 'o')

ax.annotate("Current Solution",
            xy=(-1.75, 0.8), xycoords='data', 
            xytext=(-1.95, 1.25), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle3,angleA=0,angleB=90"),
            )

ax.set_yticklabels([])
ax.set_xticklabels([])
# plt.axis('off')
plt.axis('square')
# plt.savefig('SaDe_2.png')
```

```python
delta = 0.1
x = np.arange(-2.0, 2.0, delta)
y = np.arange(-2, 2, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(0.3*(-X**2 - 3*Y**2))
# Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
# Z = (Z1 - Z2) * 2

fig, ax = plt.subplots(1, 4)
plt.rcParams["figure.figsize"] = (20, 20)
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams["axes.titleweight"] = "bold"
matplotlib.rcParams['font.size'] = 15
plt.rcParams.update({'figure.autolayout': True})

for i in range(5):
    CS = ax[i].contour(X, Y, Z1, linewidths=0.6)
    # ax.set_title('SaDe: Satisfiability Descent')
    ax[i].plot(-1.75, 0.75, 'k*')
    if i == 0:
        ax[i].annotate("Starting \nSolution",
            xy=(-1.75, 0.75), xycoords='data', 
            xytext=(-1.75, 1.75), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )

    if i == 1:
        pac = mpatches.Wedge([-1.75, 0.75], 0.5, theta1=270, theta2=360, facecolor='xkcd:grey', edgecolor='black', ls = '--', alpha=0.3)
        ax[i].add_patch(pac)
        ax[i].annotate("Solution \nRegion with gradient \ndirection [1, -1]",
            xy=(-1.75 + 0.4, 0.75 - 0.1), xycoords='data', 
            xytext=(-1.75 + 1, 1.75), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
        
    if i > 1:
        pac = mpatches.Wedge([-1.75, 0.75], 0.5, theta1=270, theta2=360, facecolor='xkcd:grey', edgecolor='black', ls = '--', alpha=0.3)
        ax[i].add_patch(pac)
        ax[i].arrow(-1.75, 0.75, 0.25, -0.3, length_includes_head=True, head_width=0.04, head_length=0.02, linewidth=2)
        if i == 2:
            ax[i].annotate("Next \nSolution",
            xy=(-1.75 + 0.25, 0.75 - 0.3), xycoords='data', 
            xytext=(-1.75 + 1, 1.75), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
    
    if i > 2:
        pac = mpatches.Wedge([-1.75 + 0.25, 0.75 - 0.3], 0.5, theta1=270, theta2=360, facecolor='xkcd:grey', edgecolor='black', ls = '--', alpha=0.3)
        ax[i].add_patch(pac)
        ax[i].arrow(-1.75 + 0.25, 0.75 - 0.3, 0.4, -0.15, length_includes_head=True, head_width=0.04, head_length=0.02, linewidth=2)
    
    if i > 3:
        pac = mpatches.Wedge([-1.75 + 0.25 + 0.4, 0.75 - 0.3 - 0.15], 0.5, theta1=270, theta2=360, facecolor='xkcd:grey', edgecolor='black', ls = '--', alpha=0.3)
        ax[i].add_patch(pac)
        ax[i].arrow(-1.75 + 0.25 + 0.4, 0.75 - 0.3 - 0.15, 0.3, -0.35, length_includes_head=True, head_width=0.04, head_length=0.02, linewidth=2)
    
    if i > 0:
        ax[i].annotate(s='', xy=(-1.75, 0.8), xytext=(-1.75 + 0.5, 0.80), arrowprops=dict(arrowstyle='<->'))
        ax[i].text(-1.75 + 0.2, 0.85, r'$\alpha$', fontsize=15)

    ax[i].set_yticklabels([])
    ax[i].set_xticklabels([])
    # plt.axis('off')
    ax[i].axis('square')
plt.savefig('SaDe_iteration.png')
```
